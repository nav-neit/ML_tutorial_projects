```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var>17].keys()
drop_columns
```

```python
df2 = df.drop(columns = drop_columns)
df2.shape
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df2.isnull())
```

```python
df3 = df2.dropna()
df3.shape
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df3.isnull())
```

```python
numeric_cols =df3.select_dtypes(include = ['int64','float64']).columns
numeric_cols
```

```python
%%capture --no-display
sns.distplot(df['MSSubClass'],hist=False,rug=True)
sns.distplot(df3['MSSubClass'],hist=False,rug=True)
```

```python
new_numeric_cols = list(numeric_cols)[1:]
print(new_numeric_cols)
len(new_numeric_cols)
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(25,25))
for i,var in enumerate(new_numeric_cols):
    plt.subplot(9,4,i+1)
    plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=1.5, wspace=0.4)
    sns.distplot(df[var],hist=False)
    sns.distplot(df3[var],hist=False) 
```

```python
categorical_cols = df3.select_dtypes(include=['object']).columns
categorical_cols
```

```python
new_cat_cols = list(categorical_cols)
new_cat_cols
```

```python
pd.concat([df[new_cat_cols[0]].value_counts()/df.shape[0]*100,
          df3[new_cat_cols[0]].value_counts()/df3.shape[0]*100],axis=1,
          keys=['MSZoning_org','MSZoning_clean'])
```

```python
def cat_var_dist(var):
    return(pd.concat([df[var].value_counts()/df.shape[0]*100,
              df3[var].value_counts()/df3.shape[0]*100],axis=1,
              keys=[var+'_org',var+'_clean']))
```

```python
for val in new_cat_cols:
    dist_df = cat_var_dist(val)
    display(dist_df)
```
