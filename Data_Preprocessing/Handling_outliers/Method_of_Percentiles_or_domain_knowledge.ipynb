### Finding Outliers

```python
import pandas as pd
```

```python
df = pd.read_csv('heights.csv')
df.head()
```

### Detecting Upper Outliers

```python
max_threshold=df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']>max_threshold]
```

### Detecting Lower Outliers

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']<min_threshold]
```

### Removing The outliers from the dataframe

```python
df_new = df[(df['height']<max_threshold) & (df['height']>min_threshold)]
df_new
```

## Outlier detection on a complex dataset

```python
df2 = pd.read_csv('bhp.csv')
display(df2.head(10))
df2.shape
```

```python
df2.describe()
```

### Min and Max Thresholds

```python
min_threshold2,max_threshold2 = df2.price_per_sqft.quantile([0.001,0.999])
print(min_threshold2,max_threshold2)
```

### Detecting lower outliers

```python
df2[df2.price_per_sqft<min_threshold2]
```

### Detecting upper outliers

```python
df2[df2.price_per_sqft>max_threshold2]
```

### Removing the outliers

```python
df2_new = df2[(df2['price_per_sqft']>min_threshold2) & (df2['price_per_sqft']<max_threshold2)]
print(df2_new.shape)
df2_new.sample(10)
```

```python
df3[df3['price_per_night']<min_threshold3]
```
